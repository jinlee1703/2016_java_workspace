### 2. OpenCV 라이브러리 Import

```python
import cv2
from google.colab.patches import cv2_imshow
```

### 3. Mean Filtering

```python
# 원본 이미지 출력
src = cv2.imread("mean_sample.png", cv2.IMREAD_GRAYSCALE)
cv2_imshow(src)

# 필터링 결과 출력
for ksize in (3, 5):
  dat = cv2.blur(src, (ksize, ksize))

  desc = 'Mean : {}x{}'.format(ksize, ksize)
  cv2.putText(dat, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, 255, 1, cv2.LINE_AA)

  cv2_imshow(dat)
  cv2.waitKey()

cv2.destroyAllWindows()
```

### 4. Median Filtering

```python
# 원본 이미지 출력
src = cv2.imread("median_sample.png", cv2.IMREAD_GRAYSCALE)
cv2_imshow(src)


# 필터링 결과 출력
dat1 = cv2.medianBlur(src, 3)
dat2 = cv2.medianBlur(src, 5)

cv2_imshow(dat1)
cv2_imshow(dat2)

cv2.waitKey()
cv2.destroyAllWindows()
```

### 5. Laplacian Filtering

```python
# 원본 이미지 출력
src = cv2.imread("laplacian_sample.png", cv2.IMREAD_GRAYSCALE)
cv2_imshow(src)

# 필터링 결과 출력
edge = cv2.Laplacian(src, -1)
cv2_imshow(edge)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
